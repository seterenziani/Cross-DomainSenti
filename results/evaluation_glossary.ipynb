```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_data(path): 
    df = pd.read_json(path, lines = True)
    return df 
```

```python
df_test1 = read_data('glossary_experiment_one.json')
df_test2 = read_data('glossary_experiment_two.json')
```

```python
plt.plot(df_test1['Glossary Size'], df_test1['Model Accuracy'], color='red', marker='o')
plt.title('Model Accuracy over Glossary Size', fontsize=12)
plt.xlabel('Glossary Size', fontsize=10)
plt.ylabel('Model Accuracy', fontsize=10)
plt.grid(True)
plt.show()
```
